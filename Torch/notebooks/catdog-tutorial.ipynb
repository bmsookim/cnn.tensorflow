```lua
require 'image';
require 'torch';
```

```lua
test_image_cat = image.load('../gen/catdog/test/8988.jpg')
test_image_dog = image.load('../gen/catdog/test/8985.jpg')

interpolation = 'bicubic'
size = 224
```

```lua
print(test_image_cat:size(2))
```

```lua
w,h = test_image_cat:size(3), test_image_cat:size(2)
```

```lua
if ((w<=h and w==size) or (h <= w and h == size)) then
  test_image_cat = test_image_cat
end
```

```lua
if w < h then
  test_image_cat = image.scale(test_image_cat, size, h/w * size, interpolation)
else
  test_image_cat = image.scale(test_image_cat, w/h * size, size, interpolation)
end
```

```lua
local w1 = math.ceil((test_image_cat:size(3) - size)/2)
local h1 = math.ceil((test_image_cat:size(2) - size)/2)
test_image_cat = image.crop(test_image_cat, w1, h1, w1+size, h1+size)
print(test_image_cat:size())
```

```lua
w,h = test_image_dog:size(3), test_image_dog:size(2)
if(w <= h and w == size) or (h <= w and h == size) then
  test_image_dog = test_image_dog
end
if w < h then
  test_image_dog = image.scale(test_image_dog, size, h/w * size, interpolation)
else
  test_image_dog = image.scale(test_image_dog, w/h * size, size, interpolation)
end

w1 = math.ceil((test_image_dog:size(3) - size)/2)
h1 = math.ceil((test_image_dog:size(2) - size)/2)
test_image_dog = image.crop(test_image_dog, w1, h1, w1+size, h1+size)
```
